The dataset used for this project contains information about Google Play Store applications. It includes 10,841 rows with details about various aspects of the apps, such as category, ratings, downloads, and more. Below is a brief explanation of the key columns:

**App Category:** The classification of each app, such as Beauty, Business, Entertainment, or Education.

**Rating:** User ratings for the app on a scale of 1 to 5, where 1 is the lowest and 5 is the highest.

**Reviews:** The total number of user reviews submitted for each app.

**Size:** The storage space required to install the app on a device.

**Installs:** The number of times each app has been downloaded by users.

**Type:** Indicates whether the app is Free or Paid.

**Price:** The cost of the app (free apps have a price of 0).

**Content Rating:** Specifies the target audience for the app—can be Everyone, Teens, or Mature audiences.

**Genres:** A more detailed sub-category for each app. For example, under Education, you may have sub-genres like Pretend Play or Language Learning.

**Last Updated:** The date of the most recent app update.

**Current Version:** The current version number of the app.

**Android Version:** The minimum Android OS version required to run the app.

**Fix the Size**

```python
 df['Size'].unique()
```

```python
def fix_size(s):
  try:
    if s.endswith('M'):
      return float(s[:-1])
    elif s.endswith('k'):
      return (float(s[:-1]) / 1024)
    elif s == '1,000+':
      return 1.0
    else:
      return np.nan
  except:
      return np.nan
```

```python
def fix_Price(A):
  try:
    if A.startswith('$'):
      return float(A[1:])
    else:
      return 0.0
  except:
    return np.nan
```

```python
df['Installs'] = df.Installs.str.replace(",","")
df['Installs'] = df.Installs.str.replace("+","")
df['Installs'] = df['Installs'].astype(float)
df['Installs'].dtype
```

```python

df['Android Ver'].unique()
```

```python
def correction(y):
  try:
    if y == 'Varies with device':
      return np.nan
    elif y == '4.4W and up':
      return y.replace('W','')
    else:
      return y
  except:
      return y
```

```python
df['Android Ver'] = df['Android Ver'].apply(correction)
```

```python
df['Android Ver'].unique()
```

```python
# Drop rows with missing ratings or crucial data in apps dataset
df.dropna(subset=['Rating'], inplace=True)

# Fill missing values for non-critical columns (like size, genre, etc.)
df['Size'].fillna(df['Size'].mode()[0], inplace=True)
df['Type'].fillna('Free', inplace=True)  # Assuming missing types are 'Free'
df['Content Rating'].fillna('Free', inplace=True)
df['Current Ver'].fillna('Free', inplace=True)
df['Android Ver'].fillna('Free', inplace=True)
```
